## Data cleaning

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
survey = pd.read_csv("survey.csv")
print(survey)
remove_nan = survey.fillna(0)
remove_nan
```

```python
remove_nan
```

```python
steps = pd.read_csv("steps.csv", sep=";")
steps.mask(steps>=60000, inplace=True)
```

```python
remove_nan2 = steps.fillna(0)
remove_nan2
```

```python
df = pd.merge(remove_nan, remove_nan2, on = 'id') 
df.head()
```

```python
df['mean'] = df.loc[:, '20-6-2013':'13-5-2014'].mean(axis = 1)
df.head()
```

After seeing the dataframe in Excel, I noticed that there are valuables that are outstanding and not credible. 

```python
def above_140(x):
    if(x > 140.0): 
        return float('0')
    else: 
        return x
df['weight'] = df['weight'].apply(above_140)

sns.histplot(df['weight'].dropna(), kde=False)
plt.title('Weight')
plt.xlabel("weight in kilo's")
plt.show()
```

## Exploratory data analysis

```python
df['mean'].name = 'Count (Steps)' #Labeling the variable for plotting
file_size = df['mean'].dropna() #Selecting the size column. I need to drop the NA's, otherwise Seaborn won't plot
sns.distplot(file_size, kde=False) 

mean_size = df['mean'].mean()
median_size = df['mean'].median()

print(f'The median filesize is: {median_size} Steps')
print(f'The mean filesize is: {mean_size} Steps')

sns.distplot(file_size, kde=False)
plt.title('Mean of steps per participant')
plt.axvline(median_size, 0, 600, color='green', label='median') #This adds a vertical line at x-position median_size, from y = 0 to y = 100 
plt.axvline(mean_size, 0, 600, color='blue', label='mean') 
plt.legend() #This adds a legend. It works automatically because I've set the labels in the previous lines
plt.show()
```

The histrogram has one peak. The mean steps of the particiapants are 7036 steps per day. There are some outliers up to 200.000 steps per day. It is unlikely that someone takes on average 200.000 steps per day. That's why I filtered those variables. The histogram has a left skew, which means that most of the results are on the left side of the histogram. 

```python
gender = sns.boxplot(x = 'gender', y = 'mean', data = df)
sns.set_style("whitegrid")
gender.set_title('Mean of steps per participant grouped by gender')
gender.set_xlabel('gender')
gender.set_ylabel('mean of steps')
```

```python
education = sns.violinplot(x = 'education', y = 'mean', data = df)
plt.title('Mean of steps per participant grouped by education')
plt.ylabel('Mean of steps')
plt.xlabel('Education')
plt.show()
```

## Lineair regression

Lineair regression is the correlation between an indepentent and dependent variables. The variable on the x-axis is independent and the variable on the y-axis is dependent. 

```python
sns.lmplot(x='weight', y='mean', data=df) #fit_reg=False leaves out the regression line; see below
plt.xlabel('Subscribers (millions)')
plt.ylabel('Views (billions)')
plt.title('YouTube channels subscribers and views')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression #we need this specific model from sk-learn

df['gender'].replace(['male', 'female'],
                        [0, 1], inplace=True)

x = df[['weight','height', 'education', 'gender']] 
y = df['mean']
lm = LinearRegression() #this creates a new LR model
lm = lm.fit(x, y) #this "fits" the model

df['mean_p'] = lm.predict(x) #use built-in predict() method to generate predictions; store in original dataframe
df.head(10)
```

## R<sup>2</sup>

```python
from sklearn.metrics import r2_score
r2_score(df['mean'], df['mean_p'])
```

The R2 is 0.01. This means that 1% of the variation in mean steps per participant can be explained. We can use the RSME to find out how many steps that are. 

## RMSE

```python
from sklearn.metrics import mean_squared_error #we need this function from sklearn
from math import sqrt #we need the square root from the math library

sqrt(mean_squared_error(df['mean'], df['mean_p']))
```

## Y-Y' plot

```python
sns.scatterplot(x='mean',y='mean_p',data=df)
plt.xlim(0, 15000) #This sets the x-axis limits to (0, 5e10 = 5000000000)
plt.ylim(0, 15000) #Ditto for y-axis. I want both axes to have the same length, so we can compare them
plt.plot([0, 15000], [0, 15000], color='red', lw=3) #This draws the straight red line, you can leave this out if you wish
plt.xlabel('Mean of steps per participant ')
plt.ylabel('Mean of steps per participant (predicted)')
plt.show()
```

On the x-axis we have the actual mean of steps per paricipant. On the y-axis we have the predicted means of steps per participant. If the model is perfect, all points should be on the line. Obviously, they are not. The actual steps per participant are not in line.
